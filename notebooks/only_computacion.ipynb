```python
df_compu_txt = pd.read_csv("data/thesis_computacion_only_txt.csv")
df_compu_txt
```

```python
from tika import parser
from progress_bar.progress_bar import printProgressBar

i = 0

def clean_spaces_text(text):
        new_text = " ".join(text.split())
        return(new_text)

def create_txt_files(path_source, path_destiny, l):
    global i
    i += 1
    printProgressBar(i, l, prefix = 'Progress:', suffix = 'Complete', length = 50)
    parsed_pdf = parser.from_file(path_source)
    data = parsed_pdf['content']
    new_data = clean_spaces_text(data)
    #print(new_data)
    
    file = open(path_destiny,"a")
    file.write(new_data)
    file.close()
```
