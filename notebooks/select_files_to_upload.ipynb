A partir de aqui se tomaran 30 tesis de Ingenieria Electrica, Computacion y Quimica. Estas tesis seran las que se utilizaran para tener las pruebas en tiempo real.

Las tesis de computacion fueron las unicas utilizadas para realizar el entrenamiento y validacion de los modelos, se incluyeron dos carreras mas por limitaciones pero para hacer probar que el clasificador de carreras funciona.

La idea es tener imagenes de las pruebas para mostrar la vista final al usuario.
